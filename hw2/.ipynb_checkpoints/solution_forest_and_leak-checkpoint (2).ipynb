```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
train = pd.read_csv("train.tsv")
test = pd.read_csv("test.tsv")
```

```python
# Сохраним ключи, которые нам понадобятся для
# сабмита в соревновние
out=test[["Num","item_id" ,"year","week","shift"]]
```

Матрица корреляции для трейна

```python
train.corr().style.background_gradient(cmap='coolwarm')
```

Матрица корреляции для теста

```python
test.corr().style.background_gradient(cmap='coolwarm')
```

Можно заметить, что переменная f%i равна f%i+30, где i=1..30, поэтому избавимся от последних тридцати переменных

```python
train=train.loc[:, train.columns[1:36]]

test=test.loc[:, test.columns[1:35]]
```

Можно заметить, что в данных имеется 53-ая неделя с аномально низкими значениями. Исключим ее.

```python
# Костыль пока избавимся от 53 недели
train=train[train["week"]!=53]
```

В данных имеется несколько ликов:

Во первых, можно восстановить значения показателей в конкретную неделю взяв будущую за нужной неделю и отступив на 1 неделю назад.

```python
# Нас интересуют данные только смещенные на одну неделю
add_0=train[train["shift"]==1]
# Предыдущий год
add_0["year"]=add_0.apply(lambda row:int(row["year"]-1) if (row["week"]==1) else int(row["year"]) , axis=1)
# Предыдущая неделя
add_0["week"]=add_0.apply(lambda row:52 if (row["week"]==1) else int(row["week"]-1) , axis=1)
#  В таком случае смещения нет
add_0["shift"]=0
# y таким образом не восстановить
add_0["y"]=np.NaN
# Добавляем к основной выборке
train=train.append(add_0)

# аналогично для теста
add_0_t=test[test["shift"]==1]
add_0_t["year"]=add_0_t.apply(lambda row:int(row["year"]-1) if (row["week"]==1) else int(row["year"]) , axis=1)
add_0_t["week"]=add_0_t.apply(lambda row:52 if (row["week"]==1) else int(row["week"]-1) , axis=1)
add_0_t["shift"]=0
add_0_t["y"]=np.NaN
test=test.append(add_0_t)
```

Перенесем значения с shift из строчек в отдельные колонки.

```python
# ключ по таблице
train=train.set_index(["shift","year","week","item_id"])
# Преенос из строчек в колонки
train=train.unstack(level="shift").reset_index()
# Назовем колонки по масе shift_фича, ключ не трогаем
train.columns = [col[0] if (col[1] == '') else str(col[1])+'_'+col[0]  for col in train.columns  ]



test=test.set_index(["shift","year","week","item_id"])
test=test.unstack(level="shift").reset_index()
test.columns = [col[0] if (col[1] == '') else str(col[1])+'_'+col[0]  for col in test.columns  ]
```

```python
# Нам нужен только один таргет
train["y"]=train["1_y"]
train=train[train["y"].isnull()==False]
train=train.drop(columns=['0_y', '1_y','2_y','3_y'])


test=test.drop(columns=['0_y', '1_y','2_y','3_y'])
# 2 недели в тесте неа самом деле не было это 
# смещенная назад 3-ая неделя. Она нам не нужна.
test=test[test["week"]!=2]
```

Было замечено, что каждую неделю происходит переход f%i-1=f%i, а f30 заменяется новым значением, причем f30 для текущей недле почти однозначно с точностью до коэффициента определяет таргет (y~0_f30*1.6). С помощью лика восстановим значение f 30 на каждую неделю и посчитаем коэффициент для каждого item_id

```python
train["coef"]=train["y"]/train["0_f30"]
only_coef=train[["coef","item_id"]]
only_coef=only_coef[only_coef["coef"].isnull()==False][only_coef["coef"]!=np.inf]
only_coef=only_coef.groupby(["item_id"]).mean()
test=pd.merge(test,only_coef,right_on=["item_id" ],left_on=["item_id" ])
```

Для строчек, где нельзя воспользоваться ликом построим простенький классификатор лес(поскольку он хорошо работает из коробки  и не требует ухещрений) Для обучения используем только 4 наиболее близкие по времени фичи

```python
train_filtered_columns=train[["y","1_f30","1_f29","1_f28","1_f27"]]
clf = RandomForestRegressor(n_estimators=500,max_depth=10,max_features=2)
clf.fit(train_filtered_columns[["1_f30","1_f29","1_f28","1_f27"]],train_filtered_columns["y"])
```

```python
# наблюдения, которые нельзя установить при помощи лика
test_5=test[test["0_f30"].isnull()==True][["year","week","item_id","1_f30","1_f29","1_f28","1_f27"]]
```

```python
# Используем лик, там где это возможно(3,4 неделя)
test_3_4=pd.DataFrame()
test_3_4['y']=test[test["0_f30"].isnull()==False]["0_f30"]*test[test["0_f30"].isnull()==False]["coef"]
test_3_4[["year","week","item_id"]]=test[test["0_f30"].isnull()==False][["year","week","item_id"]]
# Там где лик использовать нельзя используем классификатор
test_5['y']=clf.predict(test_5[["1_f30","1_f29","1_f28","1_f27"]])
```

```python
test_appended=test_3_4.append(test_5)
```

Выводим результаты в csv

```python
out=pd.merge(out,test_appended,right_on=["item_id" ,"year","week"],left_on=["item_id" ,"year","week"])
```

```python
out=out[["Num","y"]]
```
